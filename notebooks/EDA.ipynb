### How we gonna handle this data:
 - Divide the whole data into clusters 
 - In each cluster, split into train and test set
 - Apply preprocessing on each train set and just fit on test set
 - Different classification models on each cluster
 - we have two options for testing purposes: 
    - For each test data, test them on each model and then use voting classifier to get the final output
    - Or use probabilistic way to calculate each model's probability and then average them out to give the final prediction
    - Or after training get the best model and test on that only
    - Drawbacks of above mentioned point is that there's great chance that final prediction is going to be biased to one model as in the security type project test data can be of any style
 - In testing, user shall paste a link in the text field on the frontend
 - That link needs processing and data generation and on that data final prediction is going to take place
